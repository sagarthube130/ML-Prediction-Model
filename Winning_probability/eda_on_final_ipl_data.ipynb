```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
first_dataset = pd.read_csv('all_season_details.csv')
second_dataset = pd.read_csv('all_season_summary.csv')
pd.set_option('display.max_columns', None)
```

#**EDA on all_season_details**

```python
first_dataset.head()
```

```python
first_dataset.columns
```

## **1. Top Batsman by Runs**

```python
top_batsmen = first_dataset.groupby(['batsman1_name'])['runs'].sum().sort_values(ascending=False).head(10)
top_batsmen.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_batsmen.index, y=top_batsmen.values, palette='viridis',hue=top_batsmen.index,legend=False)
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.title('Top 10 Batsmen by Total Runs in IPL')
plt.xticks(rotation=45)
plt.show()
```

## **2. Top Bowlers by Wickets**

```python
top_bowlers = first_dataset.groupby(['bowler1_name'])['bowler1_wkts'].sum().sort_values(ascending=False).head(10)
top_bowlers.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_bowlers.index, y=top_bowlers.values, palette='rocket',hue=top_bowlers.index,legend=False)
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.title('Top 10 Bowlers by Total Wickets')
plt.xticks(rotation=45)
plt.show()
```

## **3.Run Distribution by Over**

```python
teams=first_dataset['home_team'].unique()
for batting_team in teams:
  temp=first_dataset[first_dataset['current_innings'] == batting_team]
  temp = temp.groupby(['over'])['runs'].sum()
  overs = first_dataset['over'].unique()
  # Sort overs to ensure the correct order
  overs_sorted = sorted(overs)
  #  distribution of total runs across different overs
  plt.figure(figsize=(10, 6))
  plt.plot(overs_sorted, temp[overs_sorted], color='skyblue', marker='o', linestyle='-', linewidth=2)

  plt.xlabel('Overs')
  plt.ylabel('Total Runs')
  plt.title(f'Total Runs Scored Across Different Overs By {batting_team}')
  plt.show()
```

## **Top Scoring teams**

```python
top_winnings = pd.DataFrame(first_dataset.groupby(['match_id','current_innings'])['runs'].sum().reset_index(),columns=['match_id','current_innings','runs'])
top_winnings = top_winnings.groupby(['current_innings'])['runs'].mean()
top_winnings_sorted = top_winnings.sort_values(ascending=False).head(10)
top_winnings_sorted.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_winnings_sorted.index, y=top_winnings_sorted.values, hue=top_winnings_sorted.index, dodge=False, legend=False)
plt.xlabel('Innings')
plt.ylabel('Average Runs')
plt.title('Top Average Runs teams per match')
plt.xticks(rotation=45)
plt.show()
```

## **Top partnership**

```python
first_dataset['partners'] = first_dataset.apply(lambda row: tuple(sorted([str(row['batsman1_name']), str(row['batsman2_name'])])), axis=1)
top_partnerships = first_dataset.groupby(['partners'])['runs'].sum().sort_values(ascending=False).head(10)
top_partnerships.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_partnerships.index.astype(str), y=top_partnerships.values, hue=top_partnerships.index.astype(str),legend=False,palette='Set2')
plt.xlabel('Batsman Pair')
plt.ylabel('Total Runs')
plt.title('Top 10 Partnerships by Total Runs')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## **Batsman vs Bowler Analysis**

```python
first_dataset['batsman_bowler'] = first_dataset['batsman1_name'] + ' vs ' + first_dataset['bowler1_name']
runs = first_dataset.groupby(['batsman_bowler'])['runs'].sum().sort_values(ascending=False).reset_index().head()
runs.head()
```

*Average runs per match*

```python
runs = first_dataset.groupby(['match_id','batsman_bowler'])['runs'].sum().sort_values(ascending=False).reset_index().head()
average_runs = runs.groupby('batsman_bowler')['runs'].mean().sort_values(ascending=False).head(10)
average_runs.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=average_runs.index, y=average_runs.values, palette='rocket',hue=average_runs.index,legend=False)
plt.xlabel('Batsman-Bowler Pair')
plt.ylabel('Average Runs')
plt.title('Top 10 Batsman-Bowler Pairs by Average Runs')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## **Top winning Teams**

```python
win_counts = second_dataset['winner'].value_counts().sort_values(ascending=False).head(10)
win_counts.head()
```

```python
plt.figure(figsize=(10, 6))
win_counts.plot(kind='bar', color='steelblue')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('Number of Wins by Each Team')
plt.xticks(rotation=45)
plt.show()
```

*Average winning Per season*

```python
temp = pd.DataFrame(second_dataset.groupby(['season', 'winner'])['winner']
                    .count()
                    .reset_index(name='count'))
temp = temp.groupby('winner')['count'].mean().reset_index(name='avg_wins_per_season')
temp['avg_wins_per_season'] = temp['avg_wins_per_season'].round(1)
temp.head()
```

```python
plt.figure(figsize=(12, 6))
plt.barh(temp['winner'], temp['avg_wins_per_season'], color='salmon')
plt.xlabel('Average Wins per Season')
plt.ylabel('Team')
plt.title('Average Wins per Season for Each Team')
plt.show()
```

## **Most Decisions**

```python
second_dataset = second_dataset[second_dataset['decision']!='no toss'].copy()
temp = second_dataset.groupby(['toss_won','decision'])['decision'].count().reset_index(name='count')

plt.figure(figsize=(10, 6))
sns.barplot(data=temp, x='decision', y='count', hue='toss_won')
plt.xlabel('Decision')
plt.ylabel('Count')
plt.title('Decision Count based on Toss Winning')
plt.show()
```

## **Impact of Toss Winning on Match Outcome**

```python
second_dataset['toss_and_match_win'] = second_dataset['toss_won'] == second_dataset['winner']

toss_match_win_count = second_dataset.groupby('toss_and_match_win').size().reset_index(name='count')

print(toss_match_win_count)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=toss_match_win_count, x='toss_and_match_win', y='count')
plt.xlabel('Toss Winner Also Won the Match')
plt.ylabel('Count')
plt.title('Counts of Toss Winners Who Also Won the Match')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

## **Identifying Captains of Winning Teams:**

```python
second_dataset['winner'].unique()
```

```python
for temp in second_dataset['winner'].unique():
  team = second_dataset[second_dataset['winner'] == temp].copy()
  team['won_captain'] = team.apply(lambda row: row['home_captain'] if row['winner'] == row['home_team'] else row['away_captain'], axis=1)
  captain_counts = team['won_captain'].value_counts().reset_index()
  captain_counts.columns = ['Captain','Win Count']

  plt.figure(figsize=(8, 6))
  sns.barplot(data=captain_counts, x='Captain', y='Win Count', palette='viridis',hue='Captain',legend=False)


  plt.xlabel('Captain')
  plt.ylabel('Win Count')
  plt.title(f'Winning Captains for Team {temp}')
  plt.xticks(rotation=45)

  plt.show()
```
